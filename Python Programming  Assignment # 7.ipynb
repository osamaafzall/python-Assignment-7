```python
class Employee:
    def __init__(self, name, number, address, salary, post, year_of_joining, education, year_of_experience):
        self.name = name
        self.number = number
        self.address = address
        self.salary = salary
        self.post = post
        self.year_of_joining = year_of_joining
        self.education = education
        self.year_of_experience = year_of_experience

    def describe_employee(self):
        print(f"Name: {self.name}\nNumber: {self.number}\nAddress: {self.address}\nSalary: {self.salary}\nPost: {self.post}\nYear of Joining: {self.year_of_joining}\nEducation: {self.education}\nYears of Experience: {self.year_of_experience}\n")

    def update_year_of_experience(self):
        self.year_of_experience += 1

    def increase_salary(self, amount):
        self.salary += amount


# Creating an instance employee101
employee101 = Employee("Muhammad osama", "101", "123 Main St", 50000, "Software Developer", 2015, "Bachelor's in CS", 8)

# Print four attributes individually
print(employee101.name)
print(employee101.number)
print(employee101.address)
print(employee101.salary)

# Call methods
employee101.describe_employee()
employee101.increase_salary(5000)
employee101.update_year_of_experience()

# Print updated details
print("\nAfter Updates:")
employee101.describe_employee()

# Creating 10 instances and calling describe_employee()
employees = [
    Employee("Haris afzal", "102", "gulshan town 13d", 60000, "Data Analyst", 2018, "Master's in Data Science", 5),
    Employee("yaqoob khan", "103", "Malir cant karachi", 55000, "Data Scientist", 2016, "Bachelor's in Statistics", 7),
    Employee("shehbaz hussain", "104", "Defence housing street.22", 70000, "AI Engineer", 2019, "Master's in AI", 4),
    Employee("zeeshan ahmed", "105", "gulistan-e-johar block 2", 65000, "ML Engineer", 2017, "PhD in Machine Learning", 6),
    Employee("daniyal nagori", "106", "Bahria town Islamabad", 48000, "Web Developer", 2020, "Bachelor's in IT", 3),
    Employee("kamran khan", "107", "punjab housing society", 72000, "Cloud Engineer", 2015, "Master's in Cloud Computing", 9),
    Employee("faisal siddiqui", "108", "scheme 33 gulistan-e-johar", 54000, "Backend Developer", 2016, "Bachelor's in Software Engineering", 6),
    Employee("farrukh haider", "109", "malir town city", 58000, "Frontend Developer", 2018, "Master's in Web Development", 5),
    Employee("saeed ahmed", "110", "model colony street.12", 50000, "Network Engineer", 2021, "Bachelor's in Network Security", 2),
    Employee("nasir buledi", "111", "pak society P.H.A", 76000, "DevOps Engineer", 2014, "Master's in DevOps", 10)
]

# Describe each employee
for emp in employees:
    emp.describe_employee()

# Apply some updates on specific employees
employees[0].increase_salary(2000)
employees[1].update_year_of_experience()
employees[2].increase_salary(1500)
```
